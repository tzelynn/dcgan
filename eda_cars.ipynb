```python
from pycocotools.coco import COCO
import os
import cv2
import matplotlib.pyplot as plt
import random
```

```python
train_path = "data/train"
coco_train = COCO(os.path.join(train_path, "COCO_train_annos.json"))
img_ids = coco_train.getImgIds()
```

```python
sample = coco_train.loadImgs(img_ids[random.randint(0, len(img_ids)-1)])[0]  # use the second image
image = cv2.imread(os.path.join(train_path, sample["file_name"]))
ann = coco_train.loadAnns(coco_train.getAnnIds(imgIds=[sample["id"]]))

for a in ann:
    x, y, w, h = a["bbox"]
    color = random.sample(range(256), 3)
    image = cv2.rectangle(image, (x, y), (x+w, y+h), color=color, thickness=3)


plt.imshow(image)
image.shape
```
